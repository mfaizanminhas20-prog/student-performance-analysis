```python
#missing values handle 
import pandas as pd

df = pd.read_csv("dirty_cafe_sales.csv")
print(df.isnull().sum())
# ERROR aur UNKNOWN ko NaN banao pehle
df.replace("ERROR", np.nan, inplace=True)
df.replace("UNKNOWN", np.nan, inplace=True)

# --------------------------------------------------------
# 1. Transaction ID
# Missing ho to drop karo — ye unique identifier hai
# Bina ID ke row ka koi matlab nahi
# --------------------------------------------------------
df.dropna(subset=["Transaction ID"], inplace=True)

# --------------------------------------------------------
# 2. Item
# Mode se fill karo — jo item sabse zyada bika
# Categorical column hai, mean/median nahi chalega
# --------------------------------------------------------
df["Item"].fillna(df["Item"].mode()[0], inplace=True)

# --------------------------------------------------------
# 3. Quantity
# Median se fill karo — numerical hai
# Median isliye kyunki outliers se affect nahi hota
# --------------------------------------------------------
df["Quantity"] = pd.to_numeric(df["Quantity"], errors='coerce')
df["Quantity"].fillna(df["Quantity"].median(), inplace=True)

# --------------------------------------------------------
# 4. Price Per Unit
# Median se fill karo — pricing mein outliers ho sakte hain
# Mean se galat average aa sakta hai
# --------------------------------------------------------
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"], errors='coerce')
df["Price Per Unit"].fillna(df["Price Per Unit"].median(), inplace=True)

# --------------------------------------------------------
# 5. Total Spent
# Mean se fill karo — normally distributed spending
# Zyada outliers nahi hain is column mein
# --------------------------------------------------------
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors='coerce')
df["Total Spent"].fillna(df["Total Spent"].mean(), inplace=True)

# --------------------------------------------------------
# 6. Payment Method
# Placeholder "Unknown" se fill karo
# Hum guess nahi kar sakte kaunsa method tha
# --------------------------------------------------------
df["Payment Method"].fillna("Unknown", inplace=True)

# --------------------------------------------------------
# 7. Location
# Placeholder "Unknown" se fill karo
# Categorical hai aur accurately guess nahi ho sakta
# --------------------------------------------------------
df["Location"].fillna("Unknown", inplace=True)

# --------------------------------------------------------
# 8. Transaction Date
# Drop karo — date ke bina transaction incomplete hai
# Time series analysis mein kaam nahi aayega
# --------------------------------------------------------
df.dropna(subset=["Transaction Date"], inplace=True)

# Final check
print("Missing values after cleaning:")
print(df.isnull().sum())
print()
print("Remaining rows:", len(df))
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("dirty_cafe_sales.csv")
df.replace(["ERROR", "UNKNOWN"], np.nan, inplace=True)
df["Total Spent"] = pd.to_numeric(df["Total Spent"], errors='coerce')
df["Total Spent"].fillna(df["Total Spent"].median(), inplace=True)

median_val = df["Total Spent"].median()

df["Transaction Label"] = df["Total Spent"].apply(
    lambda x: "High" if x > median_val else "Low"
)

print("Median:", median_val)
print(df[["Total Spent", "Transaction Label"]].head(10))
```
