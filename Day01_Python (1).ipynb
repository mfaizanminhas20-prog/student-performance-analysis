List

```python
clothes = ["Shirts", "Tie", "Pents", "shirts"]
print(clothes)       #full list
print(type(clothes))   #

print(clothes[0]) #first item

print(clothes[-1])      #last item

print(clothes[0:3])     #0-2 index printed , 3 not included

print(clothes[1:])      # start 1 end at last
print(clothes[:3])      #start 0 end at 3

# Adding items
clothes.append("jacket")        # append() adds to end
clothes.insert(1, "Jersy")      # inserts add at specific index 1 not at end
clothes[0] = "Jeans"            # change item/value at index 0 from shirts to Jeans

print(clothes)

# Removing items
clothes.remove("Tie")       # removes by a specific  value
clothes.pop(0)              # removes by specific index
print(clothes)

```

Set

```python
# Creating a set
cities={"lahore","islamabad","karachi","multan","lahore","islamabad"}

print(cities)   #prints unique set

cities.add("AB")     #add to set
print(cities)

cities.remove("AB")   #remove from set
print(cities)
```

```python
cities={"lahore","islamabad","karachi","multan","lahore","islamabad"}

print(cities[2]) # Generates Error as sets in Python are unordered collections.
```

```python
# Set operations — very powerful
web = {"Ali", "Sara", "Ahmed", "Zara"} # peeps who attended web ka bootcamp
ds = {"Ahmed", "Zara", "Bilal", "Hina"}  ## peeps who attended DS ka bootcamp

# Who is in BOTH bootcamps? we use '&'
print(web & ds)

# Who is in EITHER bootcamp?
print(web | ds)

# Who is ONLY in bootcamp A aka  web bootcamp?
print(web - ds)    # {'Ali', 'Sara'}

# Checking membership if that person belongs to that bootcamp attendees or not
print("Ali" in web)
print("Ali" in ds)

```

Dictionary

```python
student={
    "name":"manahil",
    "age": 20,
    "depar":"AI",
    "cgpa":3.82,
}
#full dictionary
print(student)

#for a specific key
print(student["age"])

#if the key-value pair is not present and we try to access it through it .get() it doesnot crash like phone is not present but doesnot give error
print(student.get("phone"))

#add new key
student["semester"] = 6
print(student)

# Removing a key
del student["cgpa"]

# Checking all keys and values
print(student.keys())         # all keys printed only
print(student.values())       # all values printed only
print(student.items())        # key-value pairs printed
```

# Notes
**List recognized by** [ ]  
* Indexes start from 0.
* Can be updated
* Items can be repeated

**Tuple recognized by** ( )
* Indexes start from 0.
* Cannot be updated Once declared, just will be used only
* Items can be repeated

**Set recognized by** { }
* No Indexes
* Can be updated
* Items are Unique. No repetition

**Dictionary recognized** { }
* Key : Value pairs
* No indexes
* Can be updated
* Keys are always unique
* Item is accesses using Key
